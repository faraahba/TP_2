```python
# Importation des bibliothèques
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lire le fichier CSV 'Shoe prices.csv' et stocker les données dans un DataFrame
shoes_dataset = pd.read_csv('Shoe prices.csv')
```

```python

# Affichage de quatre lignes aléatoires du Dataframe shoes_dataset
shoes_dataset.sample(4)
```

```python
# Suppression de la colonne 'Model' du Dataframe shoes_dataset
shoes_dataset = shoes_dataset.drop('Model', axis=1)
```

```python
# Affichage des 5 premières lignes du Dataframe shoes_dataset
shoes_dataset.head()
```

```python
# Comptage de nombre d'occurrences de chaque valeur dans la colonne 'Brand' du Dataframe shoes_dataset
shoes_dataset['Brand'].value_counts()
```

```python

# Créationd d'un diagramme en barres pour visualiser le nombre de chaussures par marque du Dataframe shoes_dataset
shoes_dataset['Brand'].value_counts().plot(kind='bar', legend='false')
```

```python
# Création d'un diagramme en barres pour visualiser le nombre de chaussures par marque du Dataframe shoes_dataset
shoes_dataset['Brand'].value_counts().plot(kind='bar', legend='false')
plt.title('Brand Counts')
plt.xlabel('Brand names')
plt.ylabel('Counts')
plt.show()
```

```python
# Conversion des valeurs de la colonne 'Type' en minuscules 
shoes_dataset['Type'] = shoes_dataset['Type'].str.lower()
```

```python
# Comptage du nombre de chaussures par type
shoes_dataset['Type'].value_counts()
```

```python
# Création d'un diagramme en barres pour visualiser le nombre de chaussures par type
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('Type Counts')
plt.xlabel('Type names')
plt.ylabel('Counts')
plt.show()
```

```python
# Définition d'une fonction pour regrouper les types de chaussures 
def add_type(inpt):
    if inpt=='casual' or inpt=='fashion' or inpt=='lifestyle' or inpt=='slides' or inpt=='retro' :
        return inpt
    else :
        return 'sport'
shoes_dataset['Type'] = shoes_dataset['Type'].apply(add_type)
```

```python
# Création d'un diagramme en barres pour visualiser le nombre d'occurrences de chaque valeur dans la colonne 'Type' du Dataframe shoes_dataset
shoes_dataset['Type'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('New Type Counts')
plt.xlabel('Type names')
plt.ylabel('Counts')
plt.show()
```

```python
# Comptage du nombre de chaussures par genre
shoes_dataset['Gender'].value_counts()
```

```python
# Création d'un diagramme circulaire pour visualiser la répartition des chaussures par genre
shoes_dataset['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle = 90)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# Remplacer toutes les occurrences de "US" par une chaîne vide dans la colonne 'Size' du Dataframe shoes_dataset
shoes_dataset['Size'] = shoes_dataset['Size'].str.replace('US','')
```

```python
# Affichage des 5 premières lignes du Dataframe shoes_dataset
shoes_dataset.head()
```

```python
# Conversion de la colonne 'Size' en type de données flottant
shoes_dataset['Size'] = shoes_dataset['Size'].astype(float)
```

```python
# Obtenir des informations sur le Dataframe shoes_dataset
shoes_dataset.info()
```

```python
# Conversion des valeurs de la colonne 'Color' en minuscules
shoes_dataset['Color'] = shoes_dataset['Color'].str.lower()
```

```python
# Comptage du nombre de chaussures par couleur
shoes_dataset['Color'].value_counts()
```

```python
# Création d'un diagramme en barres pour visualiser le nombre de chaussures par couleur
shoes_dataset['Color'].value_counts().plot(kind='bar', legend='false',  color = 'green')
plt.title('Color Counts')
plt.xlabel('Color names')
plt.ylabel('Counts')
plt.show()
```

```python
# Définition d'une fonction pour regrouper les couleurs de chaussures 
def add_Color(inpt):
    if inpt=='black' or inpt=='white' or inpt=='grey' or inpt=='black/white' or inpt=='pink' :
        return inpt
    else :
        return 'other'
shoes_dataset['Color'] = shoes_dataset['Color'].apply(add_Color)
```

```python
# Comptage du nombre d'occurrences de chaque valeur dans la colonne 'Color'
shoes_dataset['Color'].value_counts()
```

```python
# Création d'un diagramme à barres poyr visualiser le nombre d'occurrences de chaque valeur dans la colonne 'Color' du Dataframe shoes_dataset
shoes_dataset['Color'].value_counts().plot(kind='bar', legend='false',  color = 'green')
plt.title('Color Counts')
plt.xlabel('Color names')
plt.ylabel('Counts')
plt.show()
```

```python
# Conversion des valeurs de la colonne 'Material' en minuscules
shoes_dataset['Material'] = shoes_dataset['Material'].str.lower()
```

```python
# Comptage du nombre de chaussures par matériau
shoes_dataset['Material'].value_counts()
```

```python
# Création d'un diagramme en barres pour visualiser le nombre de chaussures par matériau
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('Material Counts')
plt.xlabel('Material names')
plt.ylabel('Counts')
plt.show()
```

```python
# Définition d'une fonction pour regrouper les matériaux de chaussures en deux catégories ('mech', 'leather', 'canvas', 'primeknit', 'synthetic', 'suede', 'knit', 'nylon' et 'other')
def add_Material(inpt):
    if inpt=='mech' or inpt=='leather' or inpt=='canvas' or inpt=='primeknit' or inpt=='synthetic' or inpt=='suede' or inpt=='knit' or inpt=='nylon' :
        return inpt
    else :
        return 'other'
shoes_dataset['Material'] = shoes_dataset['Material'].apply(add_Material)
```

```python
# Comptage du nombre d'occurrences de chaque valeur dans la colonne 'Material' du Dataframe shoes_dataset
shoes_dataset['Material'].value_counts()
```

```python
# Création d'un diagramme en barres pour visualiser le nombre d'occurrences de chaque valeur dans la colonne 'Material' du Dataframe shoes_dataset
shoes_dataset['Material'].value_counts().plot(kind='bar', legend='false', color = 'green')
plt.title('New Material Counts')
plt.xlabel('Material names')
plt.ylabel('Counts')
plt.show()
```

```python
# Suppression du symbole '$' des valeurs de la colonne 'Price (USD)' et conversion en type de données flottant
shoes_dataset['Price (USD)'] = shoes_dataset['Price (USD)'].str.replace('$','').astype(float)
# Affichage des 5 premières lignes du Dataframe shoes_dataset
shoes_dataset.head(5)
```

```python
# Création d'un nuage de points pour visualiser la relation entre la marque et le prix des chaussures
sns.set()
sns.scatterplot(data=shoes_dataset, x="Brand", y="Price (USD)")
```
